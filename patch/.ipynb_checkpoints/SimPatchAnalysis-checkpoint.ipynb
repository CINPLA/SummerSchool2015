**MEMBRANE RESISTANCE:** First visualize the effect of altering membrane resistance on a passive recording resulting from a 50 ms, 10 mV positive step, from a -80 mV holding potential.  To run this, modify the above code as below (be sure iter_C_m and iter_R_clamp are set to 0):

**SERIES RESISTANCE:** Next observe the effect of altering series resistance on a passive recording. Make the following modifications:

```python
params['prop'] = 'passive'
#Cycle through g_L values
params['iter_g_L'] = 0

#Cycle through R_clamp values
params['iter_R_clamp'] = 1

simulate(params)
```

You should get the following plot, which is a view of the full voltage and membrane current responses to the same step pulse performed in continuous clamp, with simulated Rs values between 10 (typical for patch clamp) and 100 MOhms (untenable for continuous patch clamp). It is also helpful to zoom in on the initial positive-going transient:

```python
params['prop'] = 'passive'
params['close_up'] = 1

simulate(params)
```

##ACTIVE PROPERTIES: 

Let’s start simply by observing the effect of altering the maximal $I_{Na}$ conductance:

```python
params['prop'] = 'active'
params['g_Na'] = 120.0 # maximum conducances, in mS/cm^2
params['g_K']  =  36.0
params['iter_g_Na'] = 1
params['close_up'] = 0
params['iter_R_clamp'] = 0 
params['R_clamp'] = 0.001 # GOhm
simulate(params)
```

This simulation clearly shows the effect of increased gNa upon measurable inward current. However, if you look carefully you can also see some slight loss of voltage control accompanying this large inward current. You can easily zoom in on that by running the script again using the close up option:

```python
params['close_up'] = 1
simulate(params)
```

This simulation was performed with a simulated series resistance of 1 MOhm (90% correction of normal $R_s$). This is close to a practical ideal. Next we see what happens when $R_s$ varies between 100 kOhms and 20 MOhms with a large inward sodium current. Run the script with:

```python
params['iter_g_Na'] = 0
params['close_up'] = 0
params['iter_R_clamp'] = 1 
#Cycle through R_c lamp values
params['R_clamps'] = [0.0001,0.001,0.002,0.005, 0.01, 0.02] #values to test
simulate(params)
```

Here you can easily see the effect that the large voltage drop across these typical values for series resistance have upon control of the clamped membrane potential and the measured sodium current.
