##Background
**Pyramidal cells:** Excitatory cell found in most areas of the cortex, recognized by the pyramidal shape of the somata.

**Stellate cells:** Any cell that have a star shaped somata, mainly inhibitory but can be excitatory.

**Basket cells:** Inhibitory interneurons with branched axonal arborizations appearing as baskets surrounding the soma of postsynaptic cells.

Interneurons can be classified by morphological, molecular, physiological and developmental data. The current standard is the Petilla terminology.

**Anatomical:** targeting pyramidal cells or displaying no target specificity and, at least in the hippocampus, those specifically targeting other interneurons. 

**Molecular:** those expressing parvalbumin (PV), including chandelier and basket cells; those expressing somatostatin (SOM), such as Martinotti cells; those expressing neuropeptide Y (NPY) but not SOM; those expressing vasoactive intestinal peptide (VIP); and those expressing cholecystokinin (CCK) but not SOM or VIP.

**Physiological:** Fast-spiking (FS) neurons show non-adapting spiking at steady-state, brief spikes and large fast after-hyperpolarizations and include continuous FS cells, delayed FS cells, stuttering FS cells and continuous stuttering FS cells. Non-adapting, non-fast spiking (NA-NFS) neurons display no apparent increase in the interspike interval at steady-state, and include continuous NA-NFS cells and burst-firing NA-NFS cells. Adapting (AD) neurons display a visually obvious increase in the interspike interval at steady-state and include continuous AD cells, bursting AD cells and delayed AD cells. Accelerating (AC) neurons display a decrease in the interspike interval at steady-state and include continuous AC cells and delayed AC cells. Irregular spiking (IS) neurons display an irregular interspike interval and include continuous IS cells and bursting IS cells. Lastly, intrinsic bursting (IB) neurons produce a stereotypical burst of two or more spikes riding on a depolarization envelope followed by a slow afterhyperpolarization potential and include rhythmic IB cells and initial IB cells, (DeFelipe et al. 2014; Petilla Interneuron Nomenclature Group, 2008).

##Exercise 2 - Firing rate analysis - a relativistic measure

```python
unit = get_unit(data['1410'])

s = unit.spiketrains[0]
f = figure(figsize=(12,8))
#raster
ax = f.add_subplot(511)
ax = raster(s, ax)
#set bin size
sigma = 1 #s
#count rate
ax = f.add_subplot(5,1,2)
nbins = s.magnitude.max() / sigma
ns, bs = histogram(s,nbins)
ax.bar(bs[0:-1], ns/sigma, width = bs[1]-bs[0])
ax.set_ylabel((qt.Hz).dimensionality)
ax.set_xticks([])
#instantaneuos rate
wins = [rectangularWindow, gaussianWindow, causalWindow]
for n, win in enumerate(wins):
    ax = f.add_subplot(5,1,n+3)
    r, t = instantaneous_rate(s, sigma, win, fs=1e3)
    ax.plot(t, r)
    ax.set_ylabel(r.dimensionality)
    if n != len(wins) - 1:
        ax.set_xticks([])
    else:
        ax.set_xlabel(t.dimensionality)
    tight_layout()
show()
```
