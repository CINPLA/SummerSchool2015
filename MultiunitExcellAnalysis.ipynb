##Background
**Pyramidal cells:** Excitatory cell found in most areas of the cortex, recognized by the pyramidal shape of the somata.

**Stellate cells:** Any cell that have a star shaped somata, mainly inhibitory but can be excitatory.

**Basket cells:** Inhibitory interneurons with branched axonal arborizations appearing as baskets surrounding the soma of postsynaptic cells.

Interneurons can be classified by morphological, molecular, physiological and developmental data. The current standard is the Petilla terminology.

**Anatomical:** targeting pyramidal cells or displaying no target specificity and, at least in the hippocampus, those specifically targeting other interneurons. 

**Molecular:** those expressing parvalbumin (PV), including chandelier and basket cells; those expressing somatostatin (SOM), such as Martinotti cells; those expressing neuropeptide Y (NPY) but not SOM; those expressing vasoactive intestinal peptide (VIP); and those expressing cholecystokinin (CCK) but not SOM or VIP.

**Physiological:** Fast-spiking (FS) neurons show non-adapting spiking at steady-state, brief spikes and large fast after-hyperpolarizations and include continuous FS cells, delayed FS cells, stuttering FS cells and continuous stuttering FS cells. Non-adapting, non-fast spiking (NA-NFS) neurons display no apparent increase in the interspike interval at steady-state, and include continuous NA-NFS cells and burst-firing NA-NFS cells. Adapting (AD) neurons display a visually obvious increase in the interspike interval at steady-state and include continuous AD cells, bursting AD cells and delayed AD cells. Accelerating (AC) neurons display a decrease in the interspike interval at steady-state and include continuous AC cells and delayed AC cells. Irregular spiking (IS) neurons display an irregular interspike interval and include continuous IS cells and bursting IS cells. Lastly, intrinsic bursting (IB) neurons produce a stereotypical burst of two or more spikes riding on a depolarization envelope followed by a slow afterhyperpolarization potential and include rhythmic IB cells and initial IB cells, (DeFelipe et al. 2014; Petilla Interneuron Nomenclature Group, 2008).

##Exercise 2: Firing rate analysis - a relativistic measure

We begin by making a raster plot where each line represents an action potential

```python
plot_spiketrain(data['1313'], 
                t_stop=5, 
                sigma=.1)
```

By counting a number of spikes $N$ in a given bin and divide on the binsize $\sigma$ we get the so called count rate $r$, i.e. $r=N/\sigma$. **Can you find the average count rate of the entire spike train?**

```python
plot_spiketrain(data['1313'], 
                t_stop=5, 
                sigma=.1, 
                histogram=True, 
                raster=False)
```

But can we actually measure the frequency of a small part of the spike train? **Try to change $\sigma$ and see if the rate changes.** Furthermore, if we have a static bin, what happens to if the bin takes only half or so of clearly separated events in the spike train? Introducing sliding windows convoluted with the spiketrains.

Imagine that we can represent each spike with a dirac $\delta$ function - zero everywhere except at the origin where it is $\infty$. Then we can define a neural response function as
$$
\rho(t) = \sum_{i=1}^N\delta(t-t_i)
$$
where $t_i$ is the timestamps of the action potentials.
We thus get the count rate by the following integral
$$
r = \frac{N}{\sigma} = \frac{1}{\sigma}\int_0^\sigma \rho(\tau) d\tau.
$$

Because the sequence of action potentials generated typically varies from bin to bin, neuronal responses are typically treated probabilistically, and characterized by the probability that a spike occurs at a particular time during a bin. Spike times are continuous variables, and as a result, the probability for a spike to occur at any moment in time is actually zero.

The firing rate $r(t)$, being a probability density, cannot be determined exactly from the limited amounts of data available from a finite number of bins. In addition, there is no unique way to approximate $r(t)$.

We may approximate the firing rate by applying a sliding window function $w$ by the following integral function
$$
r_{approx}(t) = \int_{-\infty}^\infty w(\tau) \rho(t-\tau) d\tau.
$$
Which, due to the definition of the dirac $\delta$ function, becomes
$$
r_{approx}(t) = \sum_{i=1}^N w(t-t_i).
$$

We will here present three different window functions, first the rectangular window function given by
$$
w(t) = \begin{cases} 1/\sigma & \text{if} -\sigma/2 \leq t < \sigma/2\\ 0 & \text{otherwise.} \end{cases}
$$
Giving the following rate estimation

```python
plot_spiketrain(data['1313'], 
                t_stop=5, 
                sigma=.1, 
                rectangular=True, 
                raster=False)
```

The Gaussian window function can be given by
$$
w(\tau) = \frac{1}{\sqrt{2\pi\sigma}}\exp{\left(-\frac{\tau^2}{2\sigma^2}\right)}
$$
and yield the following rate estimation

```python
plot_spiketrain(data['1313'], 
                t_stop=5, 
                sigma=.1, 
                gaussian=True, 
                raster=False)
```

Both the rectangular and the Gaussian window functions approximate the firing rate at any time using spikes fired both before and after that time. A postsynaptic neuron monitoring the spike train of a presynaptic cell only has access to spikes that have previously occurred. We thus present a causal window given as
$$
w(\tau) = \left[\frac{\tau}{\sigma^2} \exp{\left(-\frac{\tau}{\sigma}\right)} \right]_+
$$
where
$$
[z(t)]_+ = \begin{cases}z(t) & \text{if } t \geq 0 \\ 0 & \text{otherwise.} \end{cases}
$$

```python
plot_spiketrain(data['1313'],
                t_stop=5, 
                sigma=.1, 
                causal=True, 
                raster=False)
```
